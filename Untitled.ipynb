# Blood donation data by state 

```python
import pandas as pd
import matplotlib.pyplot as plt

URL_DATA = 'https://storage.data.gov.my/healthcare/blood_donations_state.parquet'

# Read the Parquet file into a DataFrame
df = pd.read_parquet(URL_DATA)

# Print all available columns in the DataFrame
print("Available columns in the DataFrame:")
print(df.columns.tolist())

print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame 'df' with the data
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])
# Filter the data to include only rows where blood_type is "all"
df_filtered = df[df['blood_type'] == 'all']

# Extract year from the date
df['year'] = df['date'].dt.year

# Create a pivot table
pivot_table = df.pivot_table(values='donations', index='year', columns='state', aggfunc='sum')

# Generate stacked bar chart
pivot_table.plot(kind='bar', stacked=True, figsize=(10, 6))

plt.xlabel('Year')
plt.ylabel('Donations')
plt.title('Donations by Year and State')
plt.legend(title='State')
plt.show()
```

```python
import seaborn as sns

# Generate heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(pivot_table, annot=True, fmt="d", cmap="YlGnBu")

plt.xlabel('State')
plt.ylabel('Year')
plt.title('Donations by Year and State')
plt.show()
```

```python


# Filter the data to include only rows where year is 2024
df_2024 = df[df['year'] == 2024]

# Group by state and sum the donations
state_donations = df_2024.groupby('state')['donations'].sum().reset_index()

# Rank the states from highest to lowest based on donations
ranked_states = state_donations.sort_values(by='donations', ascending=False).reset_index(drop=True)

# Print the ranked states
print(ranked_states)

# Visualize the ranked states
plt.figure(figsize=(10, 6))
plt.bar(ranked_states['state'], ranked_states['donations'], color='skyblue')
plt.xlabel('State')
plt.ylabel('Donations')
plt.title('Donations by State for the Year 2024')
plt.xticks(rotation=45)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have already loaded your DataFrame 'df' with the data
# Convert 'date' column to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract year from the date
df['year'] = df['date'].dt.year

# Filter the data to include only rows where year is 2024 and exclude blood_type "all"
df_2024 = df[(df['year'] == 2024) & (df['blood_type'] != 'all')]

# Create a pivot table to break down donations by state and blood_type
pivot_table = df_2024.pivot_table(values='donations', index='state', columns='blood_type', aggfunc='sum')

# Visualize the data with a grouped bar chart
pivot_table.plot(kind='bar', figsize=(10, 6))

plt.xlabel('State')
plt.ylabel('Donations')
plt.title('Donations by State and Blood Type for the Year 2024 (Excluding Blood Type "All")')
plt.legend(title='Blood Type')
plt.xticks(rotation=45)
plt.show()

```
